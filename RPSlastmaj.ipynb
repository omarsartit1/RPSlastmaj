```python
import tkinter as tk 
import random
import sys 

# Prompt choices
def play(choice):
    choices = ["rock", "paper", "scissors"]
    index = random.randint(0,2)
    computer_choice = choices[index]
    
    if choice == computer_choice :
        result.set("It's a tie !") 
    elif (choice == "rock" and computer_choice == "scissors") or \
         (choice == "scissors" and computer_choice == "paper") or \
         (choice == "paper" and computer_choice == "rock"):
        result.set("You win !")
    else :
        result.set("You lost !") 

def reset():
    reset_window = tk.Toplevel(window)
    reset_label = tk.Label(reset_window, text="Do you want to play again? (yes/no): ")
    reset_label.pack()
    reset_entry = tk.Entry(reset_window)
    reset_entry.pack()
    
    def reset_play():
        reset_choice = reset_entry.get().lower()
        if reset_choice == "yes":
            result.set("")
            reset_window.destroy()
        elif reset_choice == "no":
            Exit()
        else:
            result.set("Invalid input. Please enter 'yes' or 'no'.")
    reset_button = tk.Button(reset_window, text="Submit", command=reset_play)
    reset_button.pack()

def Exit():
    print("Thanks for playing! Goodbye!")
    window.destroy()

# GUI
window = tk.Tk()
window.title("Rock Paper Scissor")
window.configure(background='#FFA07A')  # Orange background

# Create main frames
title_frame = tk.Frame(window, bg="#FFA07A")
title_frame.pack(pady=20)

button_frame = tk.Frame(window, bg="#FFA07A")
button_frame.pack(pady=20)

reset_frame = tk.Frame(window, bg="#FFA07A")
reset_frame.pack(pady=20)

result_frame = tk.Frame(window, bg="#FFA07A")
result_frame.pack(pady=20)

# Add widgets to frames
label = tk.Label(title_frame, text="ROCK PAPER SCISSOR", foreground="white", background="#FFA07A", font=("Helvetica", 24, "bold"))
label.pack()

rock_button = tk.Button(button_frame, text="Rock", command=lambda: play("rock"))
rock_button.pack(side=tk.LEFT, padx=10)

paper_button = tk.Button(button_frame, text="Paper", command=lambda: play("paper"))
paper_button.pack(side=tk.LEFT, padx=10)

scissors_button = tk.Button(button_frame, text="Scissors", command=lambda: play("scissors"))
scissors_button.pack(side=tk.LEFT, padx=10)

reset_button = tk.Button(reset_frame, text="Reset", command=reset)
reset_button.pack()

result = tk.StringVar()
result_label = tk.Label(result_frame, textvariable=result, foreground="white", background="#FFA07A", font=("Helvetica", 18))
result_label.pack()

window.resizable(width=1000, height=1000)
window.mainloop()
```
